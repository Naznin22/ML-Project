```python
!pip install Pillow
from PIL import ImageFont, ImageDraw, Image
```

```python

def boxing(image, predictions):
    newImage = np.copy(image)

    for result_2 in predictions:
        top_x = result_2['topleft']['x']
        top_y = result_2['topleft']['y']

        btm_x = result_2['bottomright']['x']
        btm_y = result_2['bottomright']['y']

        confidence = result_2['confidence']
        label = result_2['label'] + " " + str(round(confidence, 3))

        if confidence > 0.3:
            newImage = cv2.rectangle(newImage, (top_x, top_y), (btm_x, btm_y), (255,0,0), 3)
#             newImage = cv2.putText(newImage, label, (top_x, top_y-5), cv2.FONT_HERSHEY_COMPLEX_SMALL , 0.8, (0, 230, 0), 1, cv2.LINE_AA)
          
            fontpath = "Bangla-Kolom.ttf" # 
            font = ImageFont.truetype(fontpath, 32)
            img_pil = Image.fromarray(newImage)
            draw = ImageDraw.Draw(img_pil)
            draw.text((top_x,top_y-40), label, font = font)
            newImage = np.array(img_pil)
    return newImage

```

```python
%matplotlib inline
import matplotlib.pyplot as plt

_, ax = plt.subplots(figsize=(20, 10))
ax.imshow(boxing(image, result_2))
```
